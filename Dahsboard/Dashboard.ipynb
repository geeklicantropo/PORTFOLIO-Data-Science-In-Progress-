```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv("trees.csv")
```

<h3>HISTORIGRAMAS</h3>

```python
#h = np.histogram(base.iloc[:, 1], bins = 2)
plt.hist(base.iloc[:, 1], bins = 6)
plt.title('Arvores')
plt.ylabel('Frequencia')
plt.xlabel('Altura')
```

<h3>DENSIDADE</h3>

```python
sns.distplot(base.iloc[:, 1], hist = True, kde = True,
             bins = 6, color = 'blue',
             hist_kws = {'edgecolor': 'black'})
```

<h3>DISPERSÃO</h3>

```python
plt.scatter(base.Girth, base.Volume, color = 'blue', facecolors = 'none')
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

```python
plt.plot(base.Girth, base.Volume)
```

```python
sns.regplot(base.Girth, base.Volume, data = base)
```

<h3>DISPERSÃO COM LEGENDAS</h3>

```python
base2 = pd.read_csv('co2.csv')
x = base2.conc
y = base2.uptake
```

```python
unicos = list(set(base2.Treatment))
```

```python
for i in range(len(unicos)):
    indice = base2.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label = unicos[i])
plt.legend(loc = 'lower right')
```

```python
q = base2.loc[base2['Type'] == 'Quebec']
m = base2.loc[base2['Type'] == 'Mississippi']
```
